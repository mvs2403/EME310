# Faraday's Law:
### Magnetic Flux:
 $$\Psi = \int B \cdot ds \ [Wb]$$
 
$B$ is the Magnetic Flux Density [$Wb/m^2$] and the integral is for the area through which it passes [$m^2$].
This means the Magnetic Flux is calculated with B multiplied with the Area of Loop through which it passes

### Equivalent Voltage source for emf:
$$V_f=\frac{d\Psi}{dt}$$
To get the inducaed votlage due to the magnetic flux we differentiate the Magnetic Flux with respect to time. Note that usually B is given as a function of time.


# Examples:


## Example 1:

Screenshot 2021-06-14 112528.jpg

###Magnetic Flux Density: $B = t^3 \ [Wb/m^2]$
###Magnetic Flux $\Psi [Wb]$:
$$\Psi = B \cdot \text{(area of loop)}$$

$$\Psi = t^3 \cdot (1*1.5)$$

$$\Psi = 1.5 \ t^3 \ [Wb]$$

###Equivalent emf Source:
$$V_f = \frac{d \Psi}{dt} = 4.5t^2 \ V$$

###To get $V_{total}$:
$$I = \frac{V_f}{100\Omega + 50\Omega} = \frac{4.5t^2}{150}A$$

**Note that no current flows through the Voltage meter $V$!**

$$V = I\cdot R = \frac{4.5t^2}{150}\cdot 50 = 1.5t^2 \ V$$




## Example 2:


Screenshot 2021-06-14 113224.jpg

###Magnetic Flux Density: $B = t^3 \ [Wb/m^2]$
###Magnetic Flux $\Psi [Wb]$:
$$\Psi = B \cdot \text{(area of loop)}$$

Inner Loop:
$$\Psi_1 = t^3 \cdot (1*1.5)$$

$$\Psi_1 = 1.5 \ t^3 \ [Wb]$$

Outer Loop:
$$\Psi_2 = t^3 \cdot (1*1.5)$$

$$\Psi_2 = 1.5 \ t^3 \ [Wb]$$
**Note** that the area is **ONLY** the area where the flux is not the entire area of the circuit

###Equivalent emf Source:
$$V_{f1} = \frac{d \Psi}{dt} = 4.5t^2 \ V$$
$$V_{f2} = \frac{d \Psi}{dt} = 4.5t^2 \ V$$

###To get $V_{total}$:
$$I = \frac{V_f}{100\Omega + 50\Omega} = \frac{4.5t^2}{150}A$$

**Note** that **no current** flows through the Voltage meter $V$!

$$V = V_{Outer Loop} +V_{50 \Omega}$$
$$V = V_{Outer Loop} - I\cdot 50\Omega$$
$$V = 4.5t^2 - \frac{4.5t^2}{150}\cdot 50 = -3t^2 \ V$$

**Notice** that $V_{50\Omega}$ is negative because current flows from positive to negative therfore we will minus that voltage.
